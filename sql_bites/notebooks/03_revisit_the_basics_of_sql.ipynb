### ORDER BY

```python
%%sql

SELECT *
  FROM billboard_top_100_year_end
  ORDER BY artist
  LIMIT 20;
```

In the cell below we have order the table using the artist column in alphabetical order and then ordering the year_rank in numerical order.

So we can see the artists in alphabetical order and their highest ranked songs displayed first.

```python
%%sql

SELECT *
  FROM billboard_top_100_year_end
  ORDER BY artist, year_rank
  LIMIT 30;
```

### Descending order using DESC

```python
%%sql
SELECT *
  FROM billboard_top_100_year_end
  ORDER BY artist DESC, year_rank
  LIMIT 20;
```

```python
%%sql
SELECT *
  FROM billboard_top_100_year_end
  ORDER BY artist, year_rank DESC
  LIMIT 20;
```

```python
%%sql
SELECT *
  FROM billboard_top_100_year_end
  ORDER BY artist DESC, year_rank DESC
  LIMIT 20;
```
