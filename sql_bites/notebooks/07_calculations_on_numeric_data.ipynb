### Create a new column containing the midpoint prices calculated using the high price and low price columns 

Calculate mid price - ("High Price" - "Low Price") / 2 + "Low Price"

Then add these resutls to a new column using the AS keyword

```python
%%sql

SELECT *, 
("High Price" - "Low Price") / 2 + "Low Price" AS "Midpoint Price" 
FROM us_pumpkin_prices 
LIMIT 5;
```

```python
%%sql

SELECT *,
"High Price" - "Low Price" AS "Price Range"
FROM us_pumpkin_prices 
LIMIT 5;
```

```python
%%sql

SELECT *,
"High Price" - "Low Price" AS "Price Range"
FROM us_pumpkin_prices
ORDER BY "Price Range"
LIMIT 5;
```
