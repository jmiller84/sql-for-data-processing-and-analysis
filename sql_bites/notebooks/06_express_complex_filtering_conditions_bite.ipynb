### Filtering SQL queries

### WHERE

```python
%%sql

SELECT * 
FROM billboard_top_100_year_end
WHERE year = 1984
LIMIT 5
```

### AND

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE year >= 2000
AND year_rank = 1
LIMIT 5
```

### OR

```python
%%sql

SELECT * 
FROM billboard_top_100_year_end
WHERE artist = 'Oasis'
OR artist = 'Pink Floyd'
```

### IN

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE artist IN ('Oasis', 'Pink Floyd', 'One Direction', 'Take That')
```

### LIKE

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE group_name = 'Justin Bieber'
```

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE group_name LIKE 'Justin Bieber%';
```

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE group_name LIKE '%Justin Bieber';
```

Select rows where Justin Bieber appears at the front of the group name

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE group_name LIKE '%Justin Bieber%';
```

Select rows where Justin Bieber appears at the end of the group name

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE group_name LIKE '%Justin Bieber';
```

Select all rows where Justin Bieber appears in the group name

```python
%%sql 

SELECT * 
FROM billboard_top_100_year_end
WHERE group_name LIKE '%Justin Bieber%';
```

### NOT

```python
%%sql

SELECT * 
FROM billboard_top_100_year_end
WHERE artist = 'Santana'
```

```python
%%sql

SELECT * 
FROM billboard_top_100_year_end
WHERE artist = 'Santana'
AND group_name NOT LIKE '%feat.%';
```

### BETWEEN

Note the BETWEEN keyword is inclusive meaning the result will contain bothe start adn end values within the range

```python
%%sql

SELECT * 
FROM billboard_top_100_year_end
WHERE year_rank = 1
AND year BETWEEN 1990 AND 1999
```
