## SQL Joins Exercise

Find the oldest and youngest athlets to win a medal

```python
%%sql

SELECT * 
FROM athletes 
LIMIT 10;
```

```python
%%sql

SELECT * 
FROM medals
LIMIT 10;
```

The oldest athlete to win a medal:

```python
%%sql

SELECT a.name, a.country, a.countrycode, a.sport, a.sport_code, a.date_of_birth, a.gender, m.event, m.medal
FROM athletes a
JOIN medals m
ON a.name = m.name AND a.sport = m.sport
ORDER BY date_of_birth ASC
LIMIT 10;
```

The youngest athlete to win a medal:

```python
%%sql

SELECT a.name, a.country, a.countrycode, a.sport, a.sport_code, a.date_of_birth, a.gender, m.event, m.medal
FROM athletes a
JOIN medals m
ON a.name = m.name AND a.sport = m.sport
ORDER BY date_of_birth DESC
LIMIT 10;
```

## Challenge

Use a SELF JOIN to find athletes that share a birthday.

Here we gave the same table 2 alias's a and b. 

We then joined the table with alias a and the same table with alias b using the ON conditions: 
- date of birth is the same on both tables
- the name is not the same in both tables. (otherwise we would return the whole table)

```python
%%sql

SELECT a.name, a.date_of_birth 
FROM athletes a 
JOIN athletes b
ON a.date_of_birth = b.date_of_birth AND a.name != b.name
ORDER BY date_of_birth
LIMIT 50;
```
