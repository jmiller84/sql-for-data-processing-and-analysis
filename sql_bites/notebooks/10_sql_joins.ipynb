### INNER JOIN

Inner join only returns records which are present in both tables. 

Here we are joining the athletes and medals tables, they both have a 'name' column which we will use to join the two tables on.

```python
%%sql

SELECT * 
FROM athletes 
INNER JOIN medals 
ON athletes.name = medals.name
LIMIT 20;
```

The inner join takes all of the columns for both table even if they are duplciate columns.

In the case of a duplicate column the column gets renamed e.g. country_1

```python
%%sql

SELECT * 
FROM athletes 
WHERE athletes.name = 'HOET Griet';
```

Here we can see HOET Griet competed in two events cycling road and cycling track

```python
%%sql

SELECT * 
FROM medals 
WHERE medals.name = 'HOET Griet';
```

However she only won a medal in one of the events the cycling track

```python
%%sql

SELECT * 
FROM athletes 
INNER JOIN medals 
ON athletes.name = medals.name
WHERE athletes.name = 'HOET Griet'
LIMIT 20;
```

When we join the two tables on the name column we get 2 records for HOET Griet and it appears she won a medal in both 

### Solution

```python
%%sql

SELECT * 
FROM athletes 
INNER JOIN medals 
ON athletes.name = medals.name AND athletes.sport = medals.sport
WHERE athletes.name = 'HOET Griet'
LIMIT 20;
```

Here we have joined on more than one column, now we are joining on the name and sport columns.

Meaning we will only return rows where the name and sport columns are the same in both tables

```python
%%sql

SELECT * 
FROM athletes 
INNER JOIN medals 
ON athletes.name = medals.name AND athletes.sport = medals.sport
LIMIT 20;
```

## OUTER Join

INNER JOINS only return matched records. If you want to return some unmatched records, you need an outer join. 

For example, you might want to include in your results all the athletes who didn't win any medals. 

E.g. Those athletes for whom there is no matching record in the medals table.

There are three types of OUTER JOIN

- FULL OUTER JOIN
- LEFT OUTER JOIN
- RIGHT OUTER JOIN


For the following examples, imagine that the Tokyo Paralympics are ongoing right now and that some, but not all medals have been won. 

So there would be some unmatched athletes (who have not yet won a medal) and some unmatched medals (that have not yet been won).

### FULL OUTER JOIN

The results include all matched and unmatched results

<img src='../../images/full_outer_join.png' height=300>

```python
%%sql

SELECT * 
FROM athletes 
FULL OUTER JOIN medals
ON athletes.name = medals.name
LIMIT 10;
```

### LEFT OUTER JOIN

This will return all the records from whichever table you list first in your query (i.e. on the left) and only the matched records from the second table. 

<img src='../../images/left_outer_join.png' height=300>

```python
%%sql

SELECT * 
FROM athletes 
LEFT OUTER JOIN medals
ON athletes.name = medals.name
LIMIT 10;
```

### RIGHT OUTER JOIN

This will return all the records from whichever table you list second in your query (i.e. on the left) and only the matched records from the second table.

### SELF Join

A SELF join is a regular join but the table is joined with itself. 
It is used to create a new table from matching rows within the same table.

Create a table as one was not provided

```python
%%sql

-- First, we must delete (drop) all our tables
DROP TABLE IF EXISTS staff CASCADE;
DROP SEQUENCE IF EXISTS staff_id_seq;

-- Then, we recreate them
CREATE SEQUENCE IF NOT EXISTS staff_id_seq;
CREATE TABLE staff (
    id SERIAL PRIMARY KEY,
    name VARCHAR(40),
    manager_id INT
);

-- Finally, we add any records that are needed
INSERT INTO staff (name, manager_id) VALUES ('HOET, Griet', 2);
INSERT INTO staff (name) VALUES ('HE, Zihao');
INSERT INTO staff (name, manager_id) VALUES ('AAJIM Munkhbat', 2);
INSERT INTO staff (name, manager_id) VALUES ('NEMATI Zahra', 5);
INSERT INTO staff (name) VALUES ('DRAHONINSKY David');

```

```python
%%sql

SELECT *
FROM staff;
```

Here we are selecting the rows where the id column matches the manager_id column
